```python
pd.set_option('display.max_rows', None)
data['diseases'].value_counts()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
data['diseases'].value_counts().plot(kind='hist', bins=30)
plt.title('Frequency Distribution of Diseases')
plt.xlabel('Frequency')
plt.ylabel('Number of Diseases')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
top_n = 10
top_diseases = data['diseases'].value_counts().head(top_n)
top_diseases.plot(kind='bar', color='skyblue')
plt.title(f'Top {top_n} Most Common Diseases')
plt.xlabel('Disease')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```
